## Loading modules and building your first plot

To load modules into a Python environment, simply call them via `import` statements as shown in the example below.

```python
import numpy as np
import matplotlib.pyplot as plt
```

All functions associated with a module live in what is called a *namespace*. This means that every time you want to use a function that belongs to a module, you have to add `<module_name>` in front of that function. For example, to use the `sum` function in NumPy, you need to write `numpy.sum`. The reason for using namespaces in the first place is to ensure that modules can use similar function names without those names clashing. Especially for Matplotlib, the namespace is rather long and a good programmer is inherently efficient (or lazy) and as such, you want to abbreviate these namespaces. This is where the `as` functionality comes in. The command `import matplotlib.pyplot as plt` imports the module `matplotlib.pyplot` and ensures it can be used via the abbreviated namespace `plt`.

Let me give you a nice example. Suppose you want to plot a sine, you would need to

* Define an array of numbers, i.e. the `x`-values, for which the sine needs to be evaluated.
* Calculate the value of the sine for this array of numbers, i.e. the `y`-values
* Produce a plot where you show the result.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100) # produce an array of 100 numbers between 0 and 2π at equidistant intervals
y = np.sin(x)                    # calculate the sine for each of these numbers
plt.plot(x,y)                    # plot the result
````

## Array operations and modifications

It is not uncommon to have constructed an array but wish to make a single modification to it. For example, suppose you want to have the list of numbers between 0 and 10 (exclusive), but swap the `6` for a `69`. To accomodate this, you would use the concept of **slicing** or **index-based access**.

To build a list of numbers between 0 and 10 (so excluding the 10), you would use

````python
import numpy as np

x = np.arange(0,10) # the 0 is redundant and can be ommitted, but it is sometimes good practice to emphasize upon these things
x[6] = 69           # note that the `[6]` refers to the element in the array with the index 6, it does not automagically find the number 6 in the array
print(x)
````

## Reactor modelling using NumPy and Matplotlib

Rather than providing you with a near-exhaustive list of all the functionalities of NumPy and Python, I believe it is much more worthwhile and interesting to you if you see some relevant applications pertaining to the domain of chemical engineering. In the previous quartile, you got acquainted with modelling three idealized basic reactors corresponding to the batch, continuously stirred tank reactor and the plug-flow reactor. You modelled these reactors using an analytical description of the mass-transport and kinetics. Here, we will show the numerical analogue and teach you how you can use a programming language such as Python to assist in the modelling.

Consider a plug-flow reactor for which the governing ordinary differential equation is given by

$$
v\frac{\mathrm{d} c}{\mathrm{d} x} = -kc
$$

Employing seperation of variables yields and performing discretization yields the following (strong) integral definition

$$
\int_{c_{0}}^{c_{x}} \frac{\mathrm{d}c}{c} = \int_{0}^{x} \frac{k}{v} \; \mathrm{d}x
$$

which we can readily solve to find

$$
c_{x} = c_{0} \exp \left( -\frac{k}{v} x \right).
$$

We can also numerically approximate this solution by subdividing (discretizing) the plug-flow reactor into $N$ slices wherein we assume that each slice is it own cute and tiny continuously stirred tank reactor. For these CSTR's, the following algebraic equation holds

$$
v \left(c_{i-1} - c_{i}\right) = kc_{i} \Delta x
$$

wherein the term on the left corresponds to the flow of material going into the volume and the term on the left the amount of material consumed inside the volume. Solving the above equation for $c_{i}$ yields

$$
c_{i} &= \frac{v c_{i-1}}{k \Delta x + v} \\
&=\frac{c_{i-1}}{1 + k\Delta x / v}
$$

The above equation is a so-called recursive equation. We can, recursively, calculate the value of element $i$ based on the value of element $i-1$.

Since we want to know how good our approximation is, we can introduce the following metric which quantitatively expresses this. We use the [mean absolute error](https://en.wikipedia.org/wiki/Mean_absolute_error) (MAE) for this, which is given by

$$
\eta = \frac{\sum_{i}^{N} \left| y_{i}^{\text{exact}} - y_{i}^{\text{model}} \right|}{N}
$$

This looks perhaps as a daunting equation to you, but this is in fact a very easy array equation for which we can readily use NumPy for its evaluation. The above equation can even be written a single very condensed one-liner:

```python
eta = np.average(np.abs(y_exact - y_model))
```

## Exercise 2.10

We are now in a very good position to assess the performance of our simple numerical model and explore the MAE as function of the number of element $N$. This means that we need to repeat upon our calculation. As soon as your even think about the word *repetition* in a programming context, you should immediately have a [Pavlov-reflex](https://en.wikipedia.org/wiki/Classical_conditioning#Pavlov's_research) that this implies that you need to either write a `for`-loop and embed some of the functionality inside a *function*.

To get you started, I have already build the function for you. The function lists its input and expected output.

````python
import numpy as np
import matplotlib.pyplot as plt

# set model parameters
k = 1.0          # reaction rate constant [s^-1]
v = 1.0          # superficial velocity [m/s]
c0 = 1.0         # starting concentration [mol/m^3]

def pfr_inator(k, v, c0, N):
    """
    Calculate the mean average error (MAE) determining the concentration profile of a PFR
    using N CSTRs in series.
    
    Input:
    k: reaction rate constant [s^-1]
    v: superficial velocity [m/s]
    c0: starting concentration [mol/m^3]
    N: number of CSTRs
    
    Output:
    Mean average error of the approximation
    """
    x = np.linspace(0,1,N)   # build array of x positions
    dx = x[1] - x[0]         # determine discrete interval "width of a reactor slice"
    c = np.zeros_like(x)     # initialize an array of zeros with the same dimensions as x
    c[0] = c0                # set the first value to the bulk concentration
    
    for i in range(0,N-1):
        c[i+1] = c[i] / (1 + k * dx / v)
        
    y_exact = np.exp(-k/v*x)
    eta = np.average(np.abs(y_exact - c))
    
    return eta

N_arr = np.array([2,5,10,20,50,100])
eta = np.zeros(len(N_arr)) # you might be inclined to use `zeros_like`, but that function can only be used if the type inside the array is the same

# ADD YOUR CODE HERE

# here, we are going to use a logarithmic plot as it better shows the trend we are
# interested in
plt.loglog(N_arr, eta, 'o')
plt.grid(linestyle='--')
````
